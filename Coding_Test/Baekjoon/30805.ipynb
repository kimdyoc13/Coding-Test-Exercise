```python
def solution():
    # N = int(input())
    List = list(map(int, input().split()))
    T, P = map(int, input().split())

    T_count = 0

    for i in List:
        #while        
        scissor = 1
        while i > (T * scissor):
            scissor += 1
        

        T_count += scissor
                                  
    print(T_count)
```

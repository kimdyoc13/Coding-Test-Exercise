```python
def solution():
    N = int(input())
    # 숫자 리스트로 받기
    number_list = list(map(int, input().split()))

    # answer  정의
    answer = 0
    # 리스트를 순회하면서 소수 선별
    # 2부터 자기 숫자를 제외하고 순환
    for i in number_list:
        for x in range(2, i):
            if i % x ==0 :
                print(i)
                answer +=1
                break

    print(answer)
```

```python
solution()
```

```python
def solution():
    # 입력값이 없을 때까지 반복
    while len(input())  >0:
        number_list = (list(map(int, input().split()))).sort()


        if number_list[-1]**2  == number_list[0]**2 + number_list[1]**2 :
            print("right")
        else:
            print("wrong")
```

```python
solution()
```
